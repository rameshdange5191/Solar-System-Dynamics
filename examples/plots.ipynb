Twobody Problem

Circular Orbit

```python
body0_df = pd.read_csv('outputs/body0_run_0.csv', header=None)
body1_df = pd.read_csv('outputs/body1_run_0.csv', header=None)
```

```python
display(body0_df)
```

```python
def plot_system_lab_frame():
    global body0_x, body0_y, body1_x, body1_y
    body0_x = body0_df.iloc[1:, 2]
    body0_y = body0_df.iloc[1:, 3]

    body1_x = body1_df.iloc[1:, 2]
    body1_y = body1_df.iloc[1:, 3]

    plt.axis('equal')

    plt.plot(body0_x, body0_y, label = 'Body 1')
    plt.plot(body1_x, body1_y, label = 'Body 2')

plot_system_lab_frame()
plt.title('Lab Frame')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

```python
def plot_system_com_frame():
    com_x = (1e21 * body0_x + 1e21 * body1_x) / (1e21 + 1e21)
    com_y = (1e21 * body0_y + 1e21 * body1_y) / (2e21)

    body0_com_x = body0_x - com_x
    body0_com_y = body0_y - com_y

    body1_com_x = body1_x - com_x
    body1_com_y = body1_y - com_y

    plt.axis('equal')
    plt.plot(body0_com_x, body0_com_y, label = 'Body 1')
    plt.plot(body1_com_x, body1_com_y, label = 'Body 2')
plot_system_com_frame()
plt.title('COM Frame')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

Trajectory of a photon around a black hole

```python
black_hole_df_0 = pd.read_csv('outputs/black_photon_run_0.csv', header=None)
black_hole_df_1 = pd.read_csv('outputs/black_photon_run_1.csv', header=None)
```

```python
display(black_hole_df_0)
```

```python
def plot_bh(ax):
    center = (0, 0)
    radius = 2
    circle = patches.Circle(center, radius, edgecolor='black', facecolor='black')
    ax.add_patch(circle)
    
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(black_hole_df_0.iloc[1:, 2], black_hole_df_0.iloc[1:, 3], s=1)
ax.scatter(black_hole_df_1.iloc[1:, 2], black_hole_df_1.iloc[1:, 3], s=1)
ax.set_aspect('equal')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Trajectories of two photons around a non-rotating black hole')
ax.legend(['Trajectory 1', 'Trajectory 2'])
plot_bh(ax)
plt.show()
```
